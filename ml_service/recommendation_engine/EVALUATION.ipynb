# Recommendation Engine Evaluation
This notebook trains the hybrid recommender and reports offline metrics.

```python
import os
import pandas as pd
from recommendation_engine.utils.preprocessing import load_reco_datasets
from recommendation_engine.models.hybrid import HybridRecommender
from recommendation_engine.utils.evaluation import evaluate_holdout_at_k

DATA_DIR = os.path.join(os.path.dirname(os.path.dirname(__file__)), 'datasets', 'reco')
items_df, interactions_df = load_reco_datasets(DATA_DIR)
model = HybridRecommender(alpha=0.5)
model.load_data(items_df, interactions_df)
model.train()
print('Users:', interactions_df.shape[0], 'Items:', interactions_df.shape[1])
```

```python
metrics = evaluate_holdout_at_k(model, interactions_df, k=5)
metrics
```

```python
# Show sample recommendations
sample_user = interactions_df.index[0] if len(interactions_df.index) else 'user1'
recs = model.recommend(sample_user, top_n=5)
recs
```
